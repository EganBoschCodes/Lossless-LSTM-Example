```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
true, predictions = [], []
with open('output.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for row in spamreader:
        splt = row[0].split(",")
        true.append(float(splt[0]))
        predictions.append(float(splt[1]))
```

```python
min, max = 1000, 4000

plt.plot(true[min:max], label = "True Value")
plt.plot(predictions[min:max], label = "LSTM Prediction")
```
